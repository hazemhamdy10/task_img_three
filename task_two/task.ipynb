```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
```

```python
def display_imgs(images,plots,titles,size) : 
    fig=plt.figure(figsize=size)
    for i in range (len(images))  :
        fig.add_subplot(plots[0],plots[1],i+1)
        plt.imshow(images[i],cmap='gray')
        plt.title(titles[i])
        plt.axis('off')
    plt.show()
```

```python
first_img=cv2.imread("9B6hJ.jpg")
gray=cv2.cvtColor(first_img,cv2.COLOR_BGR2GRAY)
blur=cv2.blur(gray,(7,7))
edges=cv2.Canny(blur,22,80)
contours,hier_=cv2.findContours(edges,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

for contour in contours : 
    mask=np.zeros(first_img.shape[:2],dtype=np.uint8)
    cv2.drawContours(mask,[contour],-1,255,-1)
    masked=cv2.bitwise_and(first_img,first_img,mask=mask)
    images=[mask,masked]
    titles=["mask","masked"]
    plots=[1,2]
    size=(15,15)
    print(contour)
    display_imgs(images,plots,titles,size)
    if np.sum(mask) > 0:  # Check if mask has any white pixels
        images = [mask, masked]
        titles = ["Mask ", "Masked"]
        plots = [1, 2]
        size = (15, 15)
        display_imgs(images, plots, titles, size)
    else:
        print(f"Mask for contour {contour} is empty.")

# print(edges)

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


first_img = cv2.imread("9B6hJ.jpg")
gray = cv2.cvtColor(first_img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (7, 7), 0)  
edges = cv2.Canny(blur, 22, 80)
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
for i, contour in enumerate(contours):
    if cv2.contourArea(contour)>1000 : # filtering the bad contour 
        mask = np.zeros(first_img.shape[:2], dtype=np.uint8)
        cv2.drawContours(mask, [contour], -1, 255, -1)
        masked = cv2.bitwise_and(first_img, first_img, mask=mask)
        images = [mask, masked]
        titles = [f"Mask {i+1}", f"Masked {i+1}"]
        plots = [1, 2]
        size = (15, 15)
        display_imgs(images, plots, titles, size)

 
```
