```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

1- detecting brown eye 

```python
image=cv2.imread("brown-eyes.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_copy=image.copy()
# plt.imshow(image)
# plt.show()
blur = cv2.GaussianBlur(image,(7,7),100)
# plt.imshow(blur)
# plt.show()
edges=cv2.Canny(blur,20,35)
# plt.imshow(edges,cmap='gray')
# plt.show()
Lines=cv2.HoughLinesP(edges,1,np.pi/180,15,minLineLength=1,maxLineGap=5)
Lines=Lines.reshape(-1,4)
for line in Lines : 
    x0,y0,x1,y1=line 
    cv2.line(image_copy,(x0,y0),(x1,y1),(0,0,0),3,cv2.LINE_AA)
    
plt.imshow(image_copy)
plt.show()
```

2- detecting ballon d'Or's

```python
ballon = cv2.imread("Ronaldo.jpg")
# ballon=cv2.cvtColor(ballon,cv2.COLOR_BGR2GRAY)
ballon_copy=ballon.copy()
gray= cv2.cvtColor(ballon,cv2.COLOR_BGR2GRAY)
circles=cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,dp=1,minDist=80,param1=150,param2=80,maxRadius=160,minRadius=50)
circles=circles.reshape(-1,3)
if circles is not None : 
    circles = np.uint16(np.round(circles))

    for circle in circles : 
        x,y,r = circle 
        cv2.circle(ballon_copy,(x,y),r,(255,0,0),15)

    plt.imshow(ballon_copy,cmap='gray')
    plt.show()
```
